```python
%%R
library(readr)
library(magrittr)
library(dplyr)
```

```python
%%R
data <- read.csv("drive/MyDrive/Casos_Diarios_Estado_Nacional_Defunciones_20230417.csv")
```

```python
%%R
colnames(data)
```

```python
#Promedio de defunciones ocurridas a nivel nacional en el mes de su cumpleaños durante el año 2021
```

```python
%%R
septiembre.2021 <- data %>% 
select(nombre, ends_with("09.2021")) %>% 
filter(nombre == "Nacional") %>% 
select(-nombre)

septiembre.2021 %>% 
t() %>%
mean()
```

```python
#Número total de defunciones a nivel nacional en el año 2020
```

```python
%%R
 data %>% 
select(nombre, ends_with(".2020")) %>% 
filter(nombre == "Nacional") %>% 
mutate(defunciones20 = rowSums(pick(where(is.numeric)))) %>%
select(defunciones20)
```

```python
#Mes con mayor número de defunciones a nivel nacional en el año 2022
```

```python
%%R 
resultados <- data %>% 
mutate(enero.2022 = rowSums(select(data,ends_with("01.2022"))) 
, febrero.2022 = rowSums(select(data,ends_with("02.2022"))), 
marzo.2022 = rowSums(select(data,ends_with("03.2022"))),
abril.2022 = rowSums(select(data,ends_with("04.2022"))),
mayo.2022 = rowSums(select(data,ends_with("05.2022"))),
junio.2022 = rowSums(select(data,ends_with("06.2022"))),
julio.2022 = rowSums(select(data,ends_with("07.2022"))),
agosto.2022 = rowSums(select(data,ends_with("08.2022"))),
septiembre.2022 = rowSums(select(data,ends_with("09.2022"))),
octubre.2022 = rowSums(select(data,ends_with("10.2022"))),
noviembre.2022 = rowSums(select(data,ends_with("11.2022"))),
dic.2022 = rowSums(select(data,ends_with("12.2022")))) %>% 
filter(nombre == "Nacional") %>%
select(enero.2022, febrero.2022, marzo.2022, abril.2022, mayo.2022,junio.2022,
       julio.2022,agosto.2022,septiembre.2022,octubre.2022,noviembre.2022,dic.2022) 
```

```python
%%R 
resultados %>% 
select(apply(resultados, 1,which.max))
```

```python
#Estado con el mayor número de defunciones en el mes de Agosto 2020 (sin contar la Ciudad de México ni el Estado de México)
```

```python
%%R 

data %>% 
select(nombre, ends_with("08.2020")) %>%
filter((nombre != "DISTRITO FEDERAL") & (nombre !="MEXICO" ) & (nombre != "Nacional")) %>%
mutate(num_defunciones = rowSums(pick(where(is.numeric)))) %>%
arrange(desc(num_defunciones)) %>%
select(nombre,num_defunciones) %>%
head(n=1)
```

```python
#Dia (fecha) en la que la Ciudad de México tuvo más defunciones
```

```python
%%R
res <- data %>%
filter(nombre =="DISTRITO FEDERAL") %>%
select(-nombre, -cve_ent, -poblacion)
res %>%
select(apply(res,1,which.max))
```
